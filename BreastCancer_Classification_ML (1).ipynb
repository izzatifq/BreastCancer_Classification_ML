

# **Import Library**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import datasets
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from imblearn.over_sampling import SMOTE
```

# **Data Preprocessing**


```python
# Load dataset and convert into a Dataframe
cancer = datasets.load_breast_cancer()

X = cancer.data    # input for machine learning
y = cancer.target  # ouput of machine learning

# Convert feature and target data into Dataframe
df_X = pd.DataFrame(X, columns=cancer.feature_names)
df_y = pd.Series(y, name='target')

# Combine features and targets in one Dataframes
df = pd.concat([df_X, df_y], axis=1)

df.head(10) # show 10 line of Dataframe
```

# **Exploratory Data And Analysis (EDA)**

```python
# Summary statistics of the dataset
df.describe()
```

```python
# Display dataset information
df.info()
```

```python
# Check for missing values
null_values = df.isnull().sum()
print("Jumlah nilai kosong per kolom:")
print(null_values)
```

```python
# Count target labels
print(df['target'].value_counts())
```

```python
# Standardize the features (X) using StandardScaler
scaler = StandardScaler()
scaler.fit(X)  # Fit scaler to feature data (X)
X_standardized = scaler.transform(X)  # Apply standardization to the features
```

```python
# Check class distribution before balancing
print("Class Distribution Before Balancing:\n", df['target'].value_counts())

# Using SMOTE to handle class imbalance
X = df.drop(columns=['target'])
y = df['target']
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check class distribution after balancing
print("\nClass Distribution After Balancing:\n", pd.Series(y_resampled).value_counts())
```

```python
# Check for outliers using Z-score
from scipy.stats import zscore
df_zscore = X_resampled.apply(zscore)
outliers = (df_zscore.abs() > 3).sum()
print("Number of outlier features:\n", outliers)
```

```python
# Remove outliers if they are significant
threshold = 5  # For example, if more than 5 samples on a feature are outliers, we remove them.
filtered_idx = (df_zscore.abs() <= 3).all(axis=1)
X_filtered = X_resampled[filtered_idx]
y_filtered = y_resampled[filtered_idx]

print(f"Number of samples after outlier removal: {X_filtered.shape[0]}")
```



# **Data Modelling**




```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y_filtered, test_size=0.2, random_state=42)
```

```python
# Build model (Random Forest)
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Predict on the testing data
y_pred = rf_model.predict(X_test)
```

```python
# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=cancer.target_names))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Model Accuracy
accuracy = accuracy_score(y_test, y_pred)
print('\nModel Accuracy: {0:.2f}%'.format(accuracy * 100))
```

# **Data Visualization**



```python
# Visualize the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```

```python
# Get and visualize feature importance from the model
feature_importance = rf_model.feature_importances_

# Create a DataFrame for feature importance
features = pd.Series(feature_importance, index=cancer.feature_names)
importance = features.sort_values(ascending=False)

# Visualize with a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=importance, y=importance.index, palette="coolwarm")
plt.xlabel('Feature Importance Score', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.title("Feature Importance (Random Forest)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Visualize the Correlation Matrix

plt.figure(figsize=(12, 8))
sns.heatmap(pd.concat([X_filtered, y_filtered], axis=1).corr(), cmap='coolwarm', annot=False)
plt.title("Heatmap of Correlation")
plt.show()

# View features that are most correlated with the target
correlation = pd.concat([X_filtered, y_filtered], axis=1).corr()['target'].drop('target')
print("Features with the highest correlation to the target:\n", correlation.abs().sort_values(ascending=False).head(10))
```
