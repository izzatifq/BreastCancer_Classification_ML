

# **Import Library**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import datasets
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

# **Data Preprocessing**


```python
# Load dataset and convert into a Dataframe
cancer = datasets.load_breast_cancer()

X = cancer.data    # input for machine learning
y = cancer.target  # ouput of machine learning

# Convert feature and target data into Dataframe
df_X = pd.DataFrame(X, columns=cancer.feature_names)
df_y = pd.Series(y, name='target')

# Combine features and targets in one Dataframes
df = pd.concat([df_X, df_y], axis=1)

df.head(10) # show 10 line of Dataframe
```

# **Explore Data And Analysis (EDA)**

```python
# Count target labels
print(df['target'].value_counts())
```

```python
# Summary statistics of the dataset
df.describe()
```

```python
# Display dataset information
df.info()
```

```python
# Check for missing values
null_values = df.isnull().sum()
print("Jumlah nilai kosong per kolom:")
print(null_values)
```

```python
# Check for duplicates
duplicate_count = df.duplicated().sum()
print(f"Jumlah baris duplikat: {duplicate_count}")
```

```python
# Standardize the features (X) using StandardScaler
scaler = StandardScaler()
scaler.fit(X)  # Fit scaler to feature data (X)
X_standardized = scaler.transform(X)  # Apply standardization to the features
```



# **Data Modelling**




```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.2, random_state=42)
```

```python
# Build model (Random Forest)
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
model.fit(X_train, y_train)
```

```python
# Predict on the testing data
y_pred = model.predict(X_test)
```

```python
# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=cancer.target_names))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Model Accuracy
print('\n Model Accuracy: {0:.2f}%'. format(accuracy_score(y_test, y_pred)* 100))
```

# **Data Visualization**



```python
# Visualize the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4, 3))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```

```python
# Get and visualize feature importance from the model
feature_importance = model.feature_importances_

# Create a DataFrame for feature importance
features = pd.Series(feature_importance, index=cancer.feature_names)
importance = features.sort_values(ascending=False)

# Visualize with a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=importance, y=importance.index, palette="coolwarm")
plt.xlabel('Feature Importance Score', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.title("Feature Importance (Random Forest)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Visualize the Correlation Matrix
plt.figure(figsize=(16, 10))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='Reds')
plt.title('Correlation Matrix')
plt.show()
```
